```python
df_new = df.drop(['Car','BuildingArea','YearBuilt','CouncilArea'], axis = 1)
```

```python
df_new.head()
```

```python
y = df_new['Price']
x = df_new.drop(['Price'], axis = 1)
```

```python
X_train_full, X_valid_full, y_train, y_valid = train_test_split(x, y , test_size=0.8, random_state=42)
```

```python
##low_cardinality_cols = [cname for cname in X_train_full.columns if X_train_full[cname].nunique() < 10 and X_train_full[cname].dtype in 'object']
##numerical_cols = [cname for cname in X_train_full.columns if X_train_full[cname].dtype in ['int64','float64']]
# Select categorical columns with relatively low cardinality
low_cardinality_cols = [cname for cname in X_train_full.columns if X_train_full[cname].nunique() < 10 and X_train_full[cname].dtype == 'object']

# Select numerical columns
numerical_cols = [cname for cname in X_train_full.columns if X_train_full[cname].dtype in ['int64', 'float64']]

total_cols = low_cardinality_cols + numerical_cols
total_cols
```

```python
X_train = X_train_full[total_cols].copy()
X_valid = X_valid_full[total_cols].copy()
```

```python
cat = (X_train.dtypes == 'object')
list_cat = list(cat[cat].index)
list_cat
```

```python
from sklearn.preprocessing import OneHotEncoder

OH_encoder = OneHotEncoder(handle_unknown= 'ignore', sparse_output = False)
OH_cols_train = pd.DataFrame(OH_encoder.fit_transform(X_train[list_cat]))
OH_cols_valid = pd.DataFrame(OH_encoder.fit_transform(X_valid[list_cat]))
## OH_cols_train
## OH_cols_valid

# Removing the categorical column

OH_cols_train.index = X_train.index
OH_cols_valid.index = X_valid.index

num_X_train = X_train.drop(list_cat, axis = 1)
num_X_valid = X_valid.drop(list_cat, axis = 1)

# Adding One Hot encoded columns to numerical features

OH_x_train = pd.concat([num_X_train, OH_cols_train], axis = 1)
OH_x_valid = pd.concat([num_X_valid, OH_cols_valid], axis = 1)

# Ensure that all columns have a string data

OH_x_train.columns = OH_x_train.columns.astype(str)
OH_x_valid.columns = OH_x_valid.columns.astype(str)

```

```python
data_na = (data.isnull().sum()/len(data))*100
data_na = data_na.drop(data_na[data_na == 0].index).sort_values(ascending = False)
data_na
```

```python
## 1. Dropping the data

data_dropped = data.dropna()
```

```python
## Data imutation

## Groupby method: Fill the null value with similar value

data['Engine Fuel Type'] = data.groupby('Model')['Engine Fuel Type'].transform(lambda x : x.fillna(x.mode().iloc[0] if not x.mode().empty else None))
data['Engine Fuel Type']
```

```python
data['Number of Doors'] = data.groupby('Model')['Number of Doors'].transform(lambda x : x.fillna(x.mode().iloc[0] if not x.mode().empty else None))
data['Number of Doors']
```

```python
data['Engine Cylinders'] = data.groupby('Model')['Engine Cylinders'].transform(lambda x : x.fillna(x.mean().iloc[0] if not x.mean().empty else None))
```

```python
data.info()
```

```python
data['Engine Cylinders'] = data['Engine Cylinders'].fillna(data['Engine Cylinders'].mean())
```

```python
data['Engine Cylinders']
```

```python
data['Engine HP'] = data['Engine HP'].fillna(data['Engine HP'].mean())
```

```python
data['Number of Doors'] = data['Number of Doors'].fillna(data['Number of Doors'].mean())
```
