```python
df['date'] = pd.to_datetime(df['date'])
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day
df['day_of_week'] = df['date'].dt.dayofweek
df['season'] = df['month'].apply(lambda x: 'Winter' if x in [12,1,2] else 
                               'Spring' if x in [3,4,5] else
                               'Summer' if x in [6,7,8] else 'Autumn')
```

## Temperature Trends over time

```python
plt.figure(figsize=(20, 10))
plt.subplot(3, 2, 1)
plt.plot(df['date'], df['meantemp'], alpha=0.7, linewidth=0.8)
plt.title('Daily Mean Temperature Over Time')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.grid(True, alpha=0.3)
```

```python
plt.figure(figsize=(25, 10))
plt.subplot(3, 2, 2)
monthly_temp = df.groupby(['year', 'month'])['meantemp'].mean().reset_index()
monthly_temp['date'] = pd.to_datetime(monthly_temp[['year', 'month']].assign(day=1))
plt.plot(monthly_temp['date'], monthly_temp['meantemp'], marker='o')
plt.title('Monthly Average Temperature')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.grid(True, alpha=0.3)
```

```python
plt.figure(figsize=(25, 10))
plt.subplot(3, 2, 3)
for year in df['year'].unique():
    year_data = df[df['year'] == year]
    plt.plot(year_data['date'].dt.dayofyear, year_data['meantemp'], 
             label=year, alpha=0.7)
plt.title('Temperature Patterns by Year (Day of Year)')
plt.xlabel('Day of Year')
plt.ylabel('Temperature (°C)')
plt.legend()
plt.grid(True, alpha=0.3)
```

```python
plt.figure(figsize=(25, 10))
plt.subplot(3, 2, 4)
sns.boxplot(data=df, x='season', y='meantemp')
plt.title('Temperature Distribution by Season')
plt.xticks(rotation=45)
```

```python
plt.figure(figsize=(25, 10))
plt.subplot(3, 2, 5)
yearly_temp = df.groupby('year')['meantemp'].mean()
plt.plot(yearly_temp.index, yearly_temp.values, marker='o', linewidth=2)
plt.title('Yearly Average Temperature')
plt.xlabel('Year')
plt.ylabel('Average Temperature (°C)')
plt.grid(True, alpha=0.3)
```

```python
monthly_avg_meantemp = df.groupby('month')['meantemp'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.lineplot(x='month', y='meantemp', data=monthly_avg_meantemp, marker='o')


plt.title('Average Mean Temperature by Month')
plt.xlabel('Month')
plt.ylabel('Average Mean Temperature')
plt.xticks(monthly_avg_meantemp['month']) 

plt.grid(True)
plt.show()
```

```python
monthly_yearly_avg_meantemp = df.groupby(['year', 'month'])['meantemp'].mean().reset_index()
```

```python
plt.figure(figsize=(12, 8))
sns.lineplot(
    data=monthly_yearly_avg_meantemp,
    x='month',
    y='meantemp',
    hue='year',
    marker='o',
    palette='viridis'  
)

plt.title('Average Mean Temperature by Month for Each Year (2013-2016)')
plt.xlabel('Month')
plt.ylabel('Average Mean Temperature')
plt.xticks(range(1, 13)) 
plt.legend(title='Year')

plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.show()
```

```python
plt.figure(figsize=(25, 8))

plt.subplot(2, 2, 1)
plt.plot(df['date'], df['humidity'], alpha=0.7, color='green', linewidth=0.8)
plt.title('Daily Humidity Over Time')
plt.xlabel('Date')
plt.ylabel('Humidity (%)')
plt.grid(True, alpha=0.3)
```

```python
plt.figure(figsize=(25, 8))
plt.subplot(2, 2, 2)
sns.boxplot(data=df, x='season', y='humidity')
plt.title('Humidity Distribution by Season')
plt.xticks(rotation=45)
```

```python
plt.figure(figsize=(25, 8))
plt.subplot(2, 2, 3)
plt.scatter(df['meantemp'], df['humidity'], alpha=0.5, c=df['season'].astype('category').cat.codes)
plt.xlabel('Temperature (°C)')
plt.ylabel('Humidity (%)')
plt.title('Temperature vs Humidity')
plt.colorbar(label='Season')
```

```python
plt.figure(figsize=(25, 8))

plt.subplot(2, 2, 4)
monthly_humidity = df.groupby(['year', 'month'])['humidity'].mean().reset_index()
monthly_humidity['date'] = pd.to_datetime(monthly_humidity[['year', 'month']].assign(day=1))
plt.plot(monthly_humidity['date'], monthly_humidity['humidity'], marker='o', color='green')
plt.title('Monthly Average Humidity')
plt.xlabel('Date')
plt.ylabel('Humidity (%)')
plt.grid(True, alpha=0.3)
```

## Wind speed and pressure Analysis

```python
plt.figure(figsize=(25, 10))

plt.subplot(2, 3, 1)
plt.plot(df['date'], df['wind_speed'], alpha=0.7, color='orange', linewidth=0.8)
plt.title('Wind Speed Over Time')
plt.xlabel('Date')
plt.ylabel('Wind Speed')
plt.grid(True, alpha=0.3)
```

```python
plt.figure(figsize=(25, 10))
plt.subplot(2, 3, 2)
plt.plot(df['date'], df['meanpressure'], alpha=0.7, color='red', linewidth=0.8)
plt.title('Pressure Over Time')
plt.xlabel('Date')
plt.ylabel('Pressure')
plt.grid(True, alpha=0.3)
```

```python
plt.figure(figsize=(25, 10))
plt.subplot(2, 3, 3)
sns.boxplot(data=df, x='season', y='wind_speed')
plt.title('Wind Speed by Season')
plt.xticks(rotation=45)
```

```python
plt.figure(figsize=(25, 10))
plt.subplot(2, 3, 4)
sns.boxplot(data=df, x='season', y='meanpressure')
plt.title('Pressure by Season')
plt.xticks(rotation=45)
```

```python
plt.figure(figsize=(25, 10))
plt.subplot(2, 3, 5)
# Wind speed vs pressure
plt.scatter(df['meanpressure'], df['wind_speed'], alpha=0.5)
plt.xlabel('Pressure')
plt.ylabel('Wind Speed')
plt.title('Pressure vs Wind Speed')
plt.grid(True, alpha=0.3)
```

```python
sorted_df = df.sort_values(by='wind_speed', ascending=False)
top_5_wind_days = sorted_df.head(5)
print("Top 5 days with the highest wind speed:")
print(top_5_wind_days[['date', 'wind_speed']])
```

```python
top_5_wind_days = df.sort_values(by='wind_speed', ascending=False).head(5)
top_5_wind_days['date'] = pd.to_datetime(top_5_wind_days['date'])
top_5_wind_days['date_str'] = top_5_wind_days['date'].dt.strftime('%Y-%m-%d')
plt.figure(figsize=(10, 6))
sns.barplot(x='date_str', y='wind_speed', data=top_5_wind_days, palette='viridis')

plt.title('Top 5 Days with Highest Wind Speed', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Wind Speed', fontsize=12)
plt.xticks(rotation=45) 

plt.tight_layout() 
plt.show()
```

```python
sns.lmplot(x='meantemp', y='humidity', data=df)

plt.title('Temperature vs. Humidity with Regression Line')
plt.xlabel('Mean Temperature')
plt.ylabel('Humidity')
plt.show()

plt.figure(figsize=(10, 6))
sns.scatterplot(x='meantemp', y='humidity', data=df, hue='year', palette='viridis')
plt.title('Temperature vs. Humidity, colored by Year')
plt.xlabel('Mean Temperature')
plt.ylabel('Humidity')
plt.show()
```
