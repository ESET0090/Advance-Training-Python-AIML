## Use of .loc[]

```python
import pandas as pd
df = pd.DataFrame({
  'time': ['2022-09-14 00:52:00-07:00', '2022-09-14 00:52:30-07:00',
           '2022-09-14 01:52:30-07:00'],
  'letter': ['A', 'B', 'C'],
})
df['time'] = pd.to_datetime(df.time)
df.set_index('time', inplace = True)
df.loc['2022-09-14':'2022-09-14 00:53']
```

```python
import pandas as pd
df = pd.DataFrame({
  'time': ['2022-09-14 00:52:00-07:00', '2022-09-14 00:52:30-07:00',
           '2022-09-14 01:52:30-07:00'],
  'letter': ['A', 'B', 'C'],
})
df['time'] = pd.to_datetime(df.time)

def rows_in_time_range(df, time_column, start_ts_str, timedelta_str):
    start_ts = pd.Timestamp(start_ts_str).tz_localize('US/Pacific')
    end_ts = start_ts + pd.to_timedelta(timedelta_str)

    return df.query("@start_ts <= {0} < @end_ts".format(time_column))

rows_in_time_range(df, 'time', '2022-09-14 00:00', '52 minutes 31 seconds')
```

## Pandas Dataframe: Intersect Indexes

## Pandas DataFrame: Select all rows from A that are not in B, using the index

## Pandas DataFrame: Reshape to have 1 row per value in a list column

```python
import pandas as pd
df = pd.DataFrame({
  'date': ['9/1/22', '9/2/22', '9/3/22'],
  'action': ['Add', 'Update', 'Delete'],
  'msg_ids': [[1, 2, 3], [], [2, 3]],
})
df.set_index('date', inplace=True)


temp_series = df['msg_ids'].apply(pd.Series, 1).stack()
temp_series.index = temp_series.index.droplevel(-1)
temp_series.name = 'msg_id'
new_df = temp_series.to_frame()
new_df.set_index('msg_id', inplace=True)
new_df.loc[~new_df.index.duplicated(), :] # Drop duplicates.
```

## Convert the date to datetime and extract features

```python
df['date'] = pd.to_datetime(df['date'])
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day
df['day_of_week'] = df['date'].dt.dayofweek
df['season'] = df['month'].apply(lambda x: 'Winter' if x in [12,1,2] else 
                               'Spring' if x in [3,4,5] else
                               'Summer' if x in [6,7,8] else 'Autumn')
```

```python
df
```

```python
monthly_avg_meantemp = df.groupby('month')['meantemp'].mean().reset_index()
# Step 2: Create a line plot using seaborn
plt.figure(figsize=(10, 6))
sns.lineplot(x='month', y='meantemp', data=monthly_avg_meantemp, marker='o')

# Step 3: Add labels and a title for better readability
plt.title('Average Mean Temperature by Month')
plt.xlabel('Month')
plt.ylabel('Average Mean Temperature')
plt.xticks(monthly_avg_meantemp['month']) # Ensure x-axis ticks are integer months

plt.grid(True)
plt.show()
```

```python
monthly_yearly_avg_meantemp = df.groupby(['year', 'month'])['meantemp'].mean().reset_index()
```

```python
plt.figure(figsize=(12, 8))
sns.lineplot(
    data=monthly_yearly_avg_meantemp,
    x='month',
    y='meantemp',
    hue='year',
    marker='o',
    palette='viridis'  # You can choose a different color palette
)

# Step 3: Add labels, a title, and customize the legend for better readability
plt.title('Average Mean Temperature by Month for Each Year (2013-2016)')
plt.xlabel('Month')
plt.ylabel('Average Mean Temperature')
plt.xticks(range(1, 13)) # Ensure x-axis ticks are for all 12 months
plt.legend(title='Year')

plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.show()
```
