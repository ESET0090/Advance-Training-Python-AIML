## Building a ML model from this

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import warnings
```

```python
train_df = pd.read_csv('C:\\Users\\user\\Downloads\\DailyDelhiClimateTrain.csv')
```

```python
test_df = pd.read_csv('C:\\Users\\user\\Downloads\\DailyDelhiClimateTest.csv')
```

```python
target_column = 'meantemp'
```

```python
train_df
```

```python
train_df['date'] = pd.to_datetime(train_df['date'])
train_df['year'] = train_df['date'].dt.year
train_df['month'] = train_df['date'].dt.month
train_df['day'] = train_df['date'].dt.day
train_df['day_of_week'] = train_df['date'].dt.dayofweek
train_df['season'] = train_df['month'].apply(lambda x: 'Winter' if x in [12,1,2] else 
                               'Spring' if x in [3,4,5] else
                               'Summer' if x in [6,7,8] else 'Autumn')
```

```python
train_df = train_df.drop(['date','season'],axis = 1)
```

```python
X_train = train_df.drop(target_column, axis=1)
y_train = train_df[target_column]
```

```python
test_df
```

```python
test_df.info()
```

```python
test_df['date'] = pd.to_datetime(test_df['date'])
test_df['year'] = test_df['date'].dt.year
test_df['month'] = test_df['date'].dt.month
test_df['day'] = test_df['date'].dt.day
test_df['day_of_week'] = test_df['date'].dt.dayofweek
test_df['season'] = test_df['month'].apply(lambda x: 'Winter' if x in [12,1,2] else 
                               'Spring' if x in [3,4,5] else
                               'Summer' if x in [6,7,8] else 'Autumn')
```

```python
test_df.info()
```

```python
test_df = test_df.drop(['date','season'],axis = 1)
```

```python
# Separate features and target for the testing set
X_test = test_df.drop(target_column, axis=1)
y_test = test_df[target_column]
```

```python
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
```

```python
X_test_scaled = scaler.transform(X_test)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train_scaled, y_train)
```

```python
y_pred = model.predict(X_test_scaled)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)
```

```python
r2
```

```python
print("\n--- Model Evaluation on Test Data ---")
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
print(f"R-squared (R²): {r2:.2f}")
```

```python
print("\n--- Model Coefficients ---")
print(f"Intercept: {model.intercept_:.2f}")
for i, col in enumerate(X_train.columns):
    print(f"Coefficient for {col}: {model.coef_[i]:.2f}")
```

```python
plt.scatter(y_test, y_pred, color='red', marker='o')
plt.xlabel("Actual Values (y_test)")
plt.ylabel("Predicted Values (y_pred)")
plt.title("Actual vs. Predicted Values")
plt.show()
```

```python
plt.scatter(y_test, y_pred)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='blue', linestyle='--') # Diagonal line
plt.xlabel("Actual Values (y_test)")
plt.ylabel("Predicted Values (y_pred)")
plt.title("Actual vs. Predicted Values with Ideal Line")
plt.show()
```

## Logistic Regression

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
```

```python
humidity_threshold = 85.0
train_df['is_high_humidity'] = np.where(train_df['humidity'] > humidity_threshold, 1, 0)
test_df['is_high_humidity'] = np.where(test_df['humidity'] > humidity_threshold, 1, 0)
```

```python
print(train_df[['humidity', 'is_high_humidity']].head())
```

```python
print(test_df[['humidity', 'is_high_humidity']].head())
```

```python
train_df
```

```python
test_df
```

```python
features = ['meantemp', 'wind_speed', 'meanpressure','year','month','day','day_of_week']
X_train = train_df[features]
y_train = train_df['is_high_humidity']
```

```python
X_test = test_df[features]
y_test = test_df['is_high_humidity']
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
print(f"Model coefficients: {model.coef_}")
print(f"Model intercept: {model.intercept_}")
```

```python
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
```

```python
print(f"Accuracy of the model on the test set: {accuracy:.2f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```
