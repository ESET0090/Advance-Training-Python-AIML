```python
import numpy as np
import matplotlib.pyplot as plt

# Corrected line: use np.arange (not np.arrange) and np.pi (not np)
x = np.arange(0, 3 * np.pi, 0.1)  # From 0 to 3π with step 0.1
y = np.sin(x)

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(x, y)
plt.title('Sine Wave')
plt.xlabel('x (radians)')
plt.ylabel('sin(x)')
plt.grid(True)
plt.show()
```
