## Feature Engineering

There are multiple types of feature engineering methods

1. Normalization
2. Binning

Normalization: Converting the numerical values into a std. range

Binning: Converting the numerical values into bucket of range

converting the raw data into a usable data for our model building is called as feature engg. or feature Extraction or Data Ingestion.

## Binning

1. It convert Continuous numerical values into categorical bins/intervals
2. Types of binning:

a. Equal-width: Divides the range into equal intervals

b. Equal-frequnecy: Equal no of samples per bins

c. Custom: user-defined threshold(point)

```python
conti_to_bin = ['median_income','housing_median_age']
conti_to_bin
```

```python
df['Income_bin_width']
```

```python
## Equal Width
df['Income_bin_width'] = pd.cut(df['median_income'], bins=4, labels=['Low','Medium','High','Very High'])
print(df.Income_bin_width.describe())
```

```python
df['Income_bin_width']
```

```python
## Equal Frequency
df['Income_bin_freq'] = pd.qcut(df['median_income'], q=4, labels=['Low','Medium','High','Very High'])
print(df.Income_bin_freq.describe())
```

```python
## Custom Bin
bins = [0,2,4,6,8,10]
labels = ['Very low','Low','Medium','High','Very High']
df['Income_bin_custom'] = pd.cut(df['median_income'], bins = bins, labels=labels)
print(df.Income_bin_custom.describe())
```
